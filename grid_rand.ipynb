```python
import numpy as np
import matplotlib.pyplot as plt

nx, ny = (4, 4)
x = np.linspace(-1, 1, nx+1)
y = np.linspace(-1, 1, ny+1)
xv, yv = np.meshgrid(x[:nx], y[:ny])
print(x)
print(y)

plt.plot(xv.flatten(), yv.flatten(), marker='o', color='k', linestyle='none')

for xx in x:
  plt.plot([xx,xx],[-1,1],color='k')

for yy in y:
  plt.plot([-1,1],[yy,yy],color='k')



np.random.uniform(low=-1.0, high=0.5, size=(2,))

offset_x = np.random.uniform(low=0, high=0.5, size=(nx*ny,))
offset_y = np.random.uniform(low=0, high=0.5, size=(nx*ny,))

plt.plot(offset_x+xv.flatten(),offset_y+yv.flatten(),marker='o',color='r',linestyle='none')
plt.show()

print(xv)
print(xv.flatten())
```
